## Q.1: Qual Cidade com MaiorFaturamento da Categoria 'Office Supplies'?
- Demontre o resultado através de um gráfico de barras

```python
# Criando a variável "best_seller"
best_seller = df[df["Categoria"] == "Office Supplies"].groupby("Cidade")["Valor_Venda"].sum()
```

```python
# Ordenando do maior para o menoe
best_seller = best_seller.sort_values(ascending=False)
```

```python
# Definindo os eixos X e Y
x = best_seller.index[:10]
y = best_seller.values[:10]
```

```python
# Paleta de cores
colors = ["#00004c"]

for i in range(0, 9):
    colors.append("#333333")
```

```python
# Figura, eixo
fig, ax = plt.subplots(figsize=(15, 5))

# Gráfico de barras
ax = sns.barplot(data=df, x=x, y=y, palette=colors)

# Rótulo de dados
for i in ax.patches:
    ax.annotate(f"{round(i.get_height())}", (i.get_x() + 0.4, i.get_height()),
               color="black", ha="center", va="bottom", size=13)

# Título
fig.text(x=0.13,
         y=0.91,
         s="Cidade Com Maior Faturamento Na Categoria 'Office Supplies'",
         fontdict={"size": 19,
                   "weight": "bold"},
         alpha=0.6)

# Insight
fig.text(x=0.2,
         y=0.77,
         s="$\\bf{NewYork}$ lidera o mercado\ncom mais de $\\bf{68 mil}$ em vendas.",
         fontdict={"size": 14,
                   "color": "#00004c"},
         alpha=0.7)

# Removendo os labels
plt.xticks(size=12)
plt.xlabel("")
plt.yticks([])

# Removendo as bordas
ax.spines["right"].set_visible(False)
ax.spines["left"].set_visible(False)
ax.spines["top"].set_visible(False)
ax.spines["bottom"].set_visible(False)

# Exibir
plt.show()
```

## Q.8: Considere  Que  a  Empresa  Decida  Conceder  o  Desconto  de  15%  do  Item  Anterior.  
- Qual Seria a Média do Valor de Venda Antes e Depois do Desconto?

```python
# List comprehention para aplicar os descontos de acordo com a regra de negócio
df["Desconto"] = [0.15 if x > 1000 else 0.10 for x in df["Valor_Venda"]]


# Alternativa 2
# df["Desconto"] = np.where(df["Valor_Venda"] > 1000, 0.15, 0.10)
```

```python
# Criando a variável Venda_Desconto
df["Venda_Desconto"] = df["Valor_Venda"] - df["Valor_Venda"] * df["Desconto"]
```

```python
df.head()
```

```python
# Calculando a média do valor das vendas antes do desconto
avg_before = df.loc[df["Desconto"] == 0.15, "Valor_Venda"]
print(f"A média do valor das vendas antes do desconto de 15% é U${round(avg_before.mean(), 2)}")
```

```python
# Filtrando a média das vendas após o desconto
avg_after = df.loc[df["Desconto"] == 0.15, "Venda_Desconto"]
print(f"A média do valor das vendas após o desconto de 15% seria U${round(avg_after.mean(), 2)}")
```

## Q.9:Qual o Média de Vendas Por Segmento, Por Ano e Por Mês?
- Demonstre o resultado através de gráfico de linha.

```python
# Agrupamento por Segmento, Ano e Mês em relação à média do valor das vendas
mean_challenge = df.groupby(["Ano", "Mês", "Segmento"])["Valor_Venda"].agg([np.sum, np.mean, np.median])
mean_challenge
```

```python
# Atribuindo os valores às variáveis
year = mean_challenge.index.get_level_values(0)
month = mean_challenge.index.get_level_values(1)
segment = mean_challenge.index.get_level_values(2)
```

```python
# Criação do gráfico

# Figura
plt.figure(figsize=(15, 10))

# Gráfico relacional com 4 parâmetros
fig = sns.relplot(kind = "line",
                  data = mean_challenge,
                  x = month,
                  y = "mean",
                  hue = segment, # preenchimento (linhas)
                  col = year,
                  col_wrap = 2)

plt.show()
```

## Q.10: Qual o Total de Vendas Por Categoria e SubCategoria, Considerando Somente as Top 12 SubCategorias?
- Demonstre tudo através de um único gráfico.

```python
# Número de categorias
df["SubCategoria"].nunique()
```

```python
# Verificando as SubCategorias
df["SubCategoria"].unique()
```

```python
# Agrupamento por Categoria e SubCategoria em relação à soma do valor das vendas,
# ordenando do maior para o menor faturamento e resentando o index
final_challenge = df.groupby(["Categoria", "SubCategoria"])["Valor_Venda"] \
                  .sum() \
                  .sort_values(ascending=False) \
                  .reset_index()
```

```python
# Extraindo somente os 12 maiores valores
top_12 = final_challenge[:12]
```

```python
# Ordenando por Categorias para a criação do gráfico (anel interno)
top_12 = top_12.sort_values(by="Categoria")
```

```python
top_12
```

```python
# Criando outro dataframe para o gráfico (anel exterior)
top_12_ext = top_12.groupby("Categoria")["Valor_Venda"].sum().reset_index()
top_12_ext
```

```python
# Criando a paleta de cores dos dois gráficos
catg_colors = ["#2a5d8a", "#1c3b0e", "#660000"]

sub_catg_colors = ["#698daf", "#698daf", "#698daf", "#698daf",
                   "#739f60", "#739f61", "#739f62", "#739f63",
                   "#ff4c4c", "#ff4c4d", "#ff4c4e", "#ff4c4f"]
```

```python
def pct(value):
    return f"{value:.1f}%"
```

```python
# Gráfico

# Figura, eixo
fig, ax = plt.subplots(figsize=(18,12))

# figura externa
fig1 = ax.pie(top_12_ext["Valor_Venda"],
            radius = 1,
            labels = top_12_ext["Categoria"],
            wedgeprops = dict(edgecolor = "white"),
            colors = catg_colors)

# figura interna
fig2 = ax.pie(top_12["Valor_Venda"],
              radius = 0.85,
              labels = top_12["SubCategoria"],
              wedgeprops = dict(edgecolor = "white"),
              colors = sub_catg_colors,
              autopct=lambda pct: f"${sum(top_12['Valor_Venda']):.0f}",
              labeldistance = 0.7,
              pctdistance = 0.53,
              rotatelabels = True)

# Limpa o centro do gráfico
center_circle = plt.Circle((0,0),0.5, fc = "white")

fig = plt.gcf()

# Aplicando o centro branco
fig.gca().add_artist(center_circle)

# Título
ax.text(x=-0.75,
        y=1.2,
        s="Total de Vendas de Categoria e Subcategoria (Top 12)",
        fontdict={"size": 15})

# Adicionando o total
ax.annotate(text = f"Total: ${str(int(sum(top_12['Valor_Venda'])))}" , xy = (-0.15, 0), size=12)

# Exibir
plt.show()
```
